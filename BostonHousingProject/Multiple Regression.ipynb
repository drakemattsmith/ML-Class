```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
```

```python
boston_data = load_boston()
```

```python
df = pd.DataFrame(boston_data.data, columns = boston_data.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
X = df
```

```python
y = boston_data.target
```

# Statsmodels

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
X_constant = sm.add_constant(X)
```

```python
pd.DataFrame(X_constant)
```

```python
model = sm.OLS(y, X_constant)
```

```python
lr = model.fit()
```

```python
lr.summary()
```

```python
form_lr = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', data = df)
mlr = form_lr.fit()
```

```python
mlr.summary()
```

```python
form_lr = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM', data = df)
mlr = form_lr.fit()
```

# Standardize Variable to Identify Key Feature(s)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
result = pd.DataFrame(list(zip(model.coef_, df.columns)), columns = ['coefficient', 'name']).set_index('name')
np.abs(result.sort_values(by = 'coefficient', ascending = False))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
stand_coef_linear_reg = make_pipeline(scaler, model)
```

```python
stand_coef_linear_reg.fit(X,y)
result = pd.DataFrame(list(zip(stand_coef_linear_reg.steps[1][1].coef_, df.columns)), columns = ['coefficient', 'name']).set_index('name')
```

```python
np.abs(result).sort_values(by='coefficient', ascending = False)
```

# Using R^2 to Identify Key Features

```python
from sklearn.metrics import r2_score
```

```python
lin_reg = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', data = df)
benchmark = lin_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
#Without LSTAT
lin_reg = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B', data = df)
benchmark = lin_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
#Without AGE
lin_reg = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT', data = df)
benchmark = lin_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
#Wihout DIS
lin_reg = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + RAD + TAX + PTRATIO + B + LSTAT', data = df)
benchmark = lin_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
#Without CRIM
lin_reg = smf.ols(formula = 'y ~ ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', data = df)
benchmark = lin_reg.fit()
r2_score(y, benchmark.predict(df))
```
